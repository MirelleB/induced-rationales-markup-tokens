```python
import pandas as pd

def preprocess(sentence):
  _map = {
    'I_WALK': 'WALK',
    'I_JUMP': 'JUMP',
    'I_LOOK': 'LOOK',
    'I_TURN_RIGHT': 'RIGHT',
    'I_TURN_LEFT': 'LEFT',
    'I_RUN': 'RUN'}
  tokens = [_map[token] for token in sentence.split()]
  return " ".join(tokens)

df = pd.read_csv('new_sample_collection_scan.tsv',sep='\t')
df.head()
```

```python
df['output'] = df['output'].map(preprocess)
```
