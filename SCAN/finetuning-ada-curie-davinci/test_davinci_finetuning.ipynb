```python
import pandas as pd
import json

def save_results(prompt, ground_truth, responses_, file_name):
  
  results = {'prompt':[], 'ground_truth':[],'candidate':[]}

  for p, gt, resp in zip(prompt, ground_truth, responses_):
    results['prompt'].append(p)
    results['ground_truth'].append(gt)
    results['candidate'].append(resp['choices'][0]['text'])

  #save results
  df = pd.DataFrame(results)
  df.to_csv(file_name, sep='\t', index = False)

def create_test_input(df, column):
  samples = []
  for sample in df[column]:
    samples.append(str(sample))
  return samples

def read_json(path):
  json_file = open(path, "r")
  df_ = {'prompt':[], 'completion':[]}
  for line in json_file.readlines():
    line_ = json.loads(line)
    df_['prompt'].append(line_['prompt'])
    df_['completion'].append(line_['completion'])
  return pd.DataFrame(df_)

def add_arrow(sentence):
  #new_list_sentences = []
  #for sentence in list_sentence:
  return sentence+' ->'
  #return new_list_sentences
```

```python
# Get de same test set using few-shot experiments
df_test_1_200 = pd.read_csv('scan_randon_200_test_samples.tsv',sep ='\t')
df_test_2_200 = pd.read_csv('new_sample_collection_scan.tsv',sep ='\t')
```

```python
join_dataset = {'input':df_test_1_200['instruction'].to_list()+df_test_2_200['instruction'].to_list(),
                'output':df_test_1_200['output'].to_list()+df_test_2_200['output'].to_list()}

join_dataset = pd.DataFrame(join_dataset)
join_dataset
```

```python
join_dataset['input'].to_list()[0]
```

```python
import openai
import os
test_samples = join_dataset['input'].to_list()
ground_truth_ = join_dataset['output'].to_list()
print('Len: ', len(test_samples))
print('First 5 examples: ', test_samples[0:5])
print('First 5 ground truth: ', ground_truth_[0:5])
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
responses = []

for idx, test_elem in enumerate(test_samples):
  print("Elemnt id: ", idx)
  response = openai.Completion.create(model="davinci:ft-unicamp-2022-05-17-19-03-59", prompt=test_elem, max_tokens = 256, stop="\n", temperature=0.0)
  responses.append(response)
```
