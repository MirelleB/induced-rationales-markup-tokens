## utils

```python
import string
import pandas as pd
import itertools
import random
from collections import Counter
import matplotlib.pyplot as plt
import json

def save_results(prompt, ground_truth, responses_, file_name):
  
  results = {'prompt':[], 'ground_truth':[],'candidate':[]}
  for p, gt, resp in zip(prompt, ground_truth, responses_):
    results['prompt'].append(p)
    results['ground_truth'].append(gt)
    results['candidate'].append(resp['choices'][0]['text'])

  #save results
  df = pd.DataFrame(results)
  df.to_csv(file_name, sep='\t', index = False)
```

## Prompt

```python
base_prompt =   "Example 1:\nInstruction: add the numbers 4 1 6 5 4 and 1 8 5 3\nExplanation:\nWe sum the digits, that is 4 + 3, which equals to 7. Because 7 is less than or equal to 9, this results in carry 0 and 7.\nThis carry 0 is summed with digits, that is 0 + 5 + 5, which equals to 10. Because 1 0 is greater than 9, this results in carry 1 and 0.\nThis carry 1 is summed with digits, that is 1 + 6 + 8, which equals to 15. Because 1 5 is greater than 9, this results in carry 1 and 5.\nThis carry 1 is summed with digits, that is 1 + 1 + 1, which equals to 3. Because 3 is less or equal than 9, this results in carry 0 and 3.\nThis carry 0 is summed with digits, that is 0 + 4 + 0, which equals to 4. Because 4 is less than or equal to 9, this results in carry 0 and 4. \nOutput: 4 3 5 0 7\n\nExample 2:\nInstruction: add the numbers 7 8 0 and 2 3 0 8\nExplanation:\nWe sum the digits,  that is 0 + 8, which equals to 8. Because 8 is less than or equal to 9, this results in carry 0 and 8.\nThis carry 0 is summed with digits, that is 0 + 8 + 0, which equals to 8. Because 8 is less than or equal to 9, this results in carry 0 and 8.\nThis carry 0 is summed, that is 0 + 7 + 3, which equals to 10. Because 1 0 is greater than 9, this results in carry 1 and 0.\nThis carry 1 is summed with digits, that is 1 + 0 + 2, which equals to 3. Because 3 is less than or equal to 9, this results in carry 0 and 3.\nOutput: 3 0 8 8\n\nExample 3:\nInstruction: add the numbers 9 6 4 and 9 4 2\nExplanation:\nWe sum the digits, that is 4 + 2, which equals to 6. Because 6 is less than or equal to 9, this results in carry 0 and 6.\nThis carry 0 is summed with digits, that is 0 + 6 + 4, which equals to 10. Because 1 0 is greater than 9, this results in carry 1 and 0.\nThis carry 1 is summed with digits, that is 1 + 9 + 9, which equals to 19. Because 1 9 is greater than 9, this results in carry 1 and 9.\nThis remaining carry results in 1.\nOutput: 1 9 0 6\n\nExample 4:\nInstruction: add the numbers 8 7 6 2 and 6 3 7 2\nExplanation:\nWe sum the digits, that is 2 + 2, which equals to 4. Because 4 is less than or equal to 9, this results in carry 0 and 4.\nThis carry 0 is summed with digits, that is 0 + 6 + 7, which equals to 13. Because 1 3 is greater than 9, this results in carry 1 and 3.\nThis carry 1 is summed with digits, that is 1 + 7 + 3, which equals to 11. Because 1 1 is greater than 9, this results in carry 1 and 1.\nThis carry 1 is summed with digits, that is 1 + 8 + 6, which equals to 15. Because 1 5 is greater than 9, this results in carry 1 and 5.\nThis remaining carry results in 1.\nOutput: 1 5 1 3 4\n\nExample 5:\nInstruction: add the numbers 8 0 6 2 and 2\nExplanation:\nWe sum the digits in position A, that is 2 + 2, which equals to 4. Because 4 is less than or equal to 9, this results in carry 0 and 4.\nThis carry 0 is summed with digits, that is 0 + 6 + 0, which equals to 6. Because 6 is less than or equal to 9, this results in carry 0 and 6.\nThis carry 0 is summed with digits, that is 0 + 0 + 0, which equals to 0. Because 0 is less than or equal to 9, this results in carry 0 and 0.\nThis carry 0 is summed with digits, that is 0 + 8 + 0, which equals to 8. Because 8 is less than or equal to 9, this results in carry 0 and 8.\nOutput: 8 0 6 4\n\n"
```

```python
print(base_prompt)
```

```python
test_samples = pd.read_csv(DATA_PATH, sep='\t')
```

```python
test_samples
```

```python
import os
import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
responses = []

for idx, sample in enumerate(test_samples['instruction']):

  prompt_sample=base_prompt+'Example 6:\nInstruction: '+sample+'\nExplanation:'

  response = openai.Completion.create(
  engine="text-davinci-002",
  prompt=prompt_sample,
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0,
  stop=["\n\n"]
  )
  responses.append(response)
```

```python
#save results
prompt = test_samples['instruction']
ground_truth=test_samples['output']
save_results(prompt, ground_truth, responses, SAVE_PATH)
```
